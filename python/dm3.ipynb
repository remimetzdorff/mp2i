```python
%pylab inline

def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/dm3/"+filename, bbox_inches="tight", dpi=200)
    return
```

```python
I = 0.0017
tau = 2.8e-5

def i(t):
    return I * ( 1 - np.exp(- t / (tau)) )

t = np.linspace(0,2e-4,1000)

plt.figure(figsize=(6,4))
plt.plot(t*1e6, i(t)*1e3)
plt.xlim(0,200)
plt.ylim(0,2)
plt.xlabel(r"Temps $t$ (µs)")
plt.ylabel(r"???")
plt.grid("on")

mysavefig("charge_ou_decharge.png")
```
