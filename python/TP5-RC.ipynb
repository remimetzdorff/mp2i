```python
plt.plot(portion_t,np.log(portion_u))

uy = .1 * np.ones(len(portion_u))
params, cov = np.polyfit(portion_t, np.log(portion_u),1,w=1/uy,cov=True)
a,b = params
plt.plot(portion_t, a*portion_t+b)
ua, ub = np.sqrt(diag(cov))

print("slope : {:.1f} +/- {:.1f} s^-1".format(a, ua))
tau = -1e3/a
utau = tau * np.abs(ua/a)
print("tau : {:.3f} +/- {:.3f} ms".format(tau, utau))
```
