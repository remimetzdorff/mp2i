# De l'optique géométrique au modèle des lentilles minces

```python
# figure layout
plt.figure(figsize=(16,8))
ax = plt.subplot2grid((1,1), (0,0))
ax.set_aspect("equal")
ax.set_xlim(-2.1, 10)
ax.set_ylim(-2.1, 2.1)

# PARAMÈTRES
radius = 2   # rayon du dioptre sphérique
D = 4        # diamètre du dioptre
n = 1.5      # indice de la lentille
N = 10       # nb de rayons lumineux
alpha = 0.5  # transparence rayons
frac = .6   # fraction du rayon de la lentille utilisé

# dioptre sphérique
theta = np.arcsin(D/2/radius)
t = np.linspace(-theta, theta, 1001)
x = radius * (np.cos(t) - 1)
y = radius * np.sin(t)


ax.plot([-3, 10], np.zeros(2), "k")
ax.plot(x,y,"C0")
ax.plot([np.min(x), np.min(x)], [np.min(y), np.max(y)], "C0")

# rayons
height  = np.linspace(-frac * D/2, frac * D/2, N)
for h in height:
    xd = -radius+np.sqrt(radius ** 2 - h ** 2)
    i = np.arcsin(h / radius)
    r = np.arcsin(n * h / radius)
    xf = xd + h / np.tan(r-i)
    ax.plot([-3, xd, xf, 2 * xf - xd], [h, h, 0, -h], "red", alpha=alpha)
```

# Plus plat plus près

```python
%pylab inline
```

```python
# PARAMÈTRES
# Paramètres du dioptre
R = 15        # rayon du dioptre sphérique
D = 2*R      # diamètre du dioptre
n = 1.5      # indice de la lentille
frac = .5    # fraction du rayon de la lentille utilisé
# Paramètres de tracé
alpha = 0.5  # transparence rayons
N = 10       # nb de rayons lumineux

# FIGURE LAYOUT
plt.figure(figsize=(16,12))
ax1 = plt.subplot2grid((2,1), (0,0))
ax2 = plt.subplot2grid((2,1), (1,0))
for ax in [ax1, ax2]:
    #ax.set_aspect("equal")
    ax.set_xlim(-2*R, 3*R)
    #ax.set_ylim(-2.1, 2.1)
    ax.plot([-2*R, 10], np.zeros(2), "--k", alpha=.25)
    #ax.plot([n*R/(n-1) - R], [0], "o")
    
# DIOPTRES
ouv = np.arcsin(D/2/R)
theta = np.linspace(-ouv, ouv)
# No 4P
x = R * np.cos(theta) - R
y = R * np.sin(theta)
ax1.plot([np.min(x), np.min(x)], [np.min(y), np.max(y)], "C0")
ax1.plot(x,y,"C0")
# 4P
x = -R * np.cos(theta) + np.sqrt(R**2 - D**2/4)
y = R * np.sin(theta)
ax2.plot([np.max(x), np.max(x)], [np.min(y), np.max(y)], "C0")
ax2.plot(x,y,"C0")

# RAYONS
height  = np.linspace(-frac * D/2, frac * D/2, N)
# 4P
for h in height:
    xd = -R + np.sqrt(R ** 2 - h ** 2)
    i = np.arcsin(h / R)
    r = np.arcsin(n * h / R)
    xf = xd + h / np.tan(r-i)
    ax1.plot([-2*R, xd, xf, 2 * xf - xd], [h, h, 0, -h], "red", alpha=alpha)

# 4P
for h in height:
    e = R - np.sqrt(R**2 - D**2/4)
    x = R - np.sqrt(R**2 - h**2)
    a = np.arctan(h/R)
    i1 = np.arcsin(np.sin(a)/n)
    i2 = a - i1
    i3 = np.arcsin(n*np.sin(i2))
    y = h - (e - x) * np.tan(i2)
    L = y / np.tan(i3)
    ax2.plot([-2*R, x-e, e-e, e+L-e, e+2*L-e], [h, h, y, 0, -y], "red", alpha=alpha)
```

# Condition de formation d'une image réelle par une lentille convergente

```python
def centre_image(centre_objet, f=1):
    """f>0 : lentille convergente"""
    oa = centre_objet
    of = f
    return oa * of / (oa + of)

f = 1

oa = np.linspace(-2*f, 2*f, 1001)
plt.plot(oa, centre_image(oa, f=f))
plt.ylim(-10,10)

plt.plot(oa, np.zeros(len(oa)))
```
