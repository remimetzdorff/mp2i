```python
folder = "/Users/remimetzdorff/Documents/GitHub/enseignements/2021-2022_cpge_paul_valery_paris/tp/tp5/"

data = np.genfromtxt(folder+"scope_0.csv", delimiter=',')
t = data[2:,0]
e = data[2:,1]
u = data[2:,2]
N = len(t)

plt.plot(t,e)
plt.plot(t,u)

i_start = 500
i_stop  = 600
portion_t = t[i_start:i_stop]
portion_u = u[i_start:i_stop]
plt.plot(portion_t,portion_u)
```

```python
plt.plot(portion_t,np.log(portion_u))

params, cov = np.polyfit(portion_t, np.log(portion_u),1,cov=True)
a,b = params
plt.plot(portion_t, a*portion_t+b)
ua, ub = np.sqrt(diag(cov))

print("slope : {:.1f} +/- {:.1f} s^-1".format(a, ua))
tau = -1e3/a
utau = tau * np.abs(ua/a)
print("tau : {:.1f} +/- {:.1f} ms".format(tau, utau))
```

```python
x = np.linspace(0,1,1000)
# comment and uncomment the last term to see how the fit appears in the figure,
# and how the covariances of the single polynomial coefficients vary in turn.
y = np.cos(x) * x**2 + x + np.sin(x - 1.) \
#     + (x * 1.3)**6

p, cov = np.polyfit(x, y, 2, cov=True)

plt.plot(x, y)
plt.plot(x, np.polyval(p,x))
plt.show()

print(np.sqrt(np.diag(cov)))
```
