```python
%pylab qt

def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/chap4/"+filename, bbox_inches="tight", dpi=200)
    return

def lowpass(t, y, fc):
    Te = t[1]-t[0]
    tau = 1/(2*np.pi*fc)
    filtered_y = [0]
    for i, val in enumerate(y):
        if i==0:
            pass
        else:
            a = tau/Te
            filtered_y.append((a/(1+a))*filtered_y[-1]+1/(1+a) * val)
    return np.array(filtered_y)
```

```python
# Décharge du condensateur

E = 1500 # tension du générateur haute tension en volts
C = 470e-6 # capacité du condensateur en farads
R = 50 # résistance du thorax du patient en ohms
W = 400 # énergie du choc en joules

def uc(t):
    return E * np.exp(- t / (R*C))

t = np.linspace(0,30e-3,1000)
u = uc(t)

t1 = - R * C * np.log(np.sqrt( 1 - 2 * W / C / E ** 2))

u_vraie = np.array([max(uc(t1),val) for val in u])

plt.figure(figsize=(6,4))
plt.plot(t*1e3, u*1e-3, "--C0", alpha=.25)
plt.plot(t*1e3, u_vraie*1e-3, "C0")
plt.xlim(0,30)
plt.ylim(0,1.6)
plt.xlabel(r"Temps $t$ (ms)")
plt.ylabel(r"Tension $u_C(t)$ (kV)")
plt.grid("on")

mysavefig("defibrillateur_choc.png")
```

```python
# Charge du condensateur

E = 1500 # tension du générateur haute tension en volts
C = 470e-6 # capacité du condensateur en farads
r = 1e3 # résistance du circuit de charge en ohm

def uc(t):
    return E * ( 1 - np.exp(- t / (r*C)) )

t = np.linspace(0,3,1000)
u = uc(t)

plt.figure(figsize=(6,4))
plt.plot(t, np.ones(len(t))*E*1e-3, "--C1", alpha=.25)
plt.plot(t, u*1e-3, "C1")
plt.xlim(0,3)
plt.ylim(0,1.6)
plt.xlabel(r"Temps $t$ (s)")
plt.ylabel(r"Tension $u_C(t)$ (kV)")
plt.grid("on")

mysavefig("defibrillateur_charge.png")
```

```python
# RC with square as input
t  = np.linspace(0,2,1000)
u1 = np.sign(np.sin(2*np.pi*t))
u2 = lowpass(t,u1,3)

plt.figure(figsize=(4,3))
plt.plot(t,u1, label=r"$e$")
plt.plot(t,u2, label=r"$u_C$")
plt.xlabel(r"$t$ (s)")
plt.ylabel(r"$u$ (V)")
plt.legend()

mysavefig("experience_RC.png")
```

```python
# Charge du condensateur (TD)

E = 5 # tension du générateur haute tension en volts
C = 470e-9 # capacité du condensateur en farads
r = 100 # résistance du circuit de charge en ohm

def uc(t):
    return E * ( 1 - np.exp(- t / (r*C)) )

t = np.linspace(0,.4e-3,1000)
u = uc(t)

plt.figure(figsize=(6,4))
plt.plot(t*1e6, u)
plt.xlim(0,400)
plt.ylim(0,6)
plt.xlabel(r"Temps $t$ ($\mu$s)")
plt.ylabel(r"Tension aux bornes du condensateur (V)")
plt.grid("on")

mysavefig("rc_charge_td.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#############################
# PARAMÈTRES DE LA RÉSOLUTION
#############################
t0 = 0                  # bornes de l'intervalle de résolution
tf = 1 
dt = 1e-2                # pas de temps
n  = int((tf-t0)/dt + 1) # nombre de points
t = np.linspace(t0,tf,n) # temps en seconde

##########################
# PARAMÈTRES DU CIRCUIT RC
##########################
u0  = 1   # CI en volts
tau = .1 # temps caractéristique en secondes

############
# SIGNAL GBF
############
E   = 0   # offset
amp = 1   # amplitude
f   = 3   # fréquence en hertz

constant = np.ones(n) * E
step     = np.array([(E if val>=0 else 0) for val in t]) 
square   = amp * np.sign(np.sin(2*np.pi*f*t)) + E
sinus    = amp * np.sin(2*np.pi*f*t) + E
spike    = np.array([(E if val>=0 and val<t[-1]/20 else 0) for val in t]) 
noise    = np.random.normal(0,.005,n) # modélisation d'un bruit aléatoire

e = step + noise # signal aux bornes du GBF
# À vous de créer le signal que vous voulez tester...

#################
# MÉTHODE D'EULER
#################
u = np.zeros(len(t))  # préparation du tableau
u[0] = u0             # initialisation en fonction de la CI
for i in range(n-1):  # méthode d'Euler explicite
    u[i+1] = e[i] * dt / tau + u[i] * (1 - dt / tau)

##########################
# REPRÉSENTATION GRAPHIQUE
##########################
plt.plot(t,e, label="e(t)")
plt.plot(t,u, label="u(t)")
plt.xlabel("Temps (s)")
plt.ylabel("Tension (V)")

uc_decharge = u0 * np.exp(-t/tau) # formule analytique
plt.plot(t,uc_decharge, label="théo")
plt.legend()
```
