```python
%pylab inline

def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/dm2/"+filename, bbox_inches="tight", dpi=200)
    return
```

```python
# Charge du condensateur (TD)

E = 15e3 # tension du générateur haute tension en volts
tau = 2e-3

def uc(t):
    return E * np.exp(- t / tau)

t = np.linspace(0,6e-3,1000)
u = uc(t)

plt.figure(figsize=(6,4))
plt.plot(t*1e3, u*1e-3)
plt.xlim(0,6)
plt.ylim(0,16)
plt.xlabel(r"Temps $t$ (ms)")
plt.ylabel(r"Tension aux bornes de la bougie $u_2(t)$ (V)")
plt.grid("on")

mysavefig("tension_bougie.png")
```
