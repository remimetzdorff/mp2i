```python
%pylab inline
def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/colles/"+filename, bbox_inches="tight", dpi=600)
    return
```

```python
# exo18 electrocin√©tique
R = 4e3
L = 0.2
em = 8
omega  = 2*np.pi*1/1.25e-3
omega0 = R/2/L 

plt.figure(figsize=(6,4))
t = np.linspace(0,3,1000)*1e-3

e = em * np.cos(omega*t)
h = 1 / (2-2j*omega0/omega)
sm = em * np.abs(h)
phi = np.angle(h)
s = sm * np.cos(omega*t + phi)

plt.plot(t*1e3,e)
plt.plot(t*1e3,s)
plt.grid()
plt.xlim(min(t)*1e3, max(t)*1e3)
plt.xlabel("Temps (ms)")
plt.ylabel("Tension (V)")

mysavefig("exo18.pdf")
```
