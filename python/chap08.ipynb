```python
%pylab inline

def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/chap8/"+filename, bbox_inches="tight", dpi=600)
    return

def spring(r,k=1,r0=1):
    return k * (r-r0)**2 / 2
```

# Oscillateur de Landau

```python
def landau(x, k=1, l0=1, a=1):
    l = np.sqrt(a**2 + x**2)
    return 1 / 2 * k  * (l - l0)**2

x = np.linspace(-3,3,1000)
for a in [2,1,.1,.01]:
    plt.plot(x,landau(x,a=a))
```

```python
a = .5
xeq = np.sqrt(1-a**2)

plt.figure(figsize=(4,3))
ax = plt.subplot2grid((1,1), (0,0))
x = np.linspace(-3,3,1000)
plt.plot(x, landau(x, a=a))
plt.xlim(-2,2)
plt.ylim(-.05,.6)
plt.xlabel("$x$")
plt.ylabel("${\cal E}_\mathrm{p}$")
plt.plot([-xeq, -xeq], [-1,1], "--C1")
plt.plot([0, 0], [-1,1], "--C1")
plt.plot([xeq, xeq], [-1,1], "--C1")
ax.set_xticks([-2*xeq, -xeq, 0, xeq, 2*xeq])
ax.set_xticklabels(["$-2x_\mathrm{éq}$", "$-x_\mathrm{éq}$", 0, "$x_\mathrm{éq}$", "$2x_\mathrm{éq}$"])
mysavefig("landau.pdf")
```

# Pendule simple

```python
m = 1
g = 1
l = 1
theta = np.linspace(-np.pi,np.pi)

def Ep(theta):
    return m * g * l * (1 - np.cos(theta))

k = m*g*l

plt.plot(theta, Ep(theta))
plt.plot(theta, spring(theta,k=k,r0=0))

plt.xlim(-np.pi, np.pi)
plt.ylim()
```

# Potentiel de Lennard Jones

```python
E0 = 1
d  = 1
r = np.linspace(1e-1*d, 5*d,1000)

def lennardjones(r, E0=1, d=1):
    return 4 * E0 * ( (d/r)**12 - (d/r)**6 )

def ddlennardjones(r, E0=1, d=1):
    return 4 * E0 * ( 12*13*d**12/r**14 - 6*7*d**6/r**8 )

plt.plot(r, lennardjones(r,E0=E0,d=d)/E0)

plt.ylim(-1,5)
plt.xlim(0.5,1.5)

req = d * 2 **(1/6)
k = ddlennardjones(req, E0=E0, d=d)
plt.plot(r, spring(r,k=k,r0=req) + lennardjones(req,E0=E0,d=d))
```

```python
plt.figure(figsize=(6,4))
plt.plot(r, lennardjones(r,E0=E0,d=d)/E0)
plt.ylim(-1.5,2.5)
plt.xlim(0.5,3)
plt.xlabel("Distance interatomique $r/r_0$")
plt.ylabel("Énergie potentielle ${\cal E}_\mathrm{p}/{\cal E}_0$")
plt.title("Potentiel de Lennard-Jones")
mysavefig("lennard_jones.pdf")
```

# Pour le TD

```python
def gaussian(x,x0=0,w=1,scale=1,offset=0):
    return scale * np.exp(-(x-x0)**2/2/w**2) + offset

def Ep(x, E0):
    return E0 * (gaussian(x) + gaussian(x,w=.5,scale=-1.25))

x = np.linspace(-5,5,1000)

plt.figure(figsize=(6,4))
plt.plot(x, Ep(x,1), "C2")
plt.grid(True)
plt.xlabel("$x$ (m)")
plt.ylabel("${\cal E}_p$ (J)")
plt.title("Profil d'énergie potentielle")
plt.xlim(-5,5)
mysavefig("exo4.png")
```
