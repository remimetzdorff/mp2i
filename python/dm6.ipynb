```python
%pylab inline

def mysavefig(filename):
    plt.tight_layout()
    plt.savefig("../images/dm6/"+filename, bbox_inches="tight", dpi=600)
    return
```

```python
def xi(x, eta):
    return -eta*np.cos(x) + 2 * (np.cos(x/2) - .4)**2

x = np.linspace(-np.pi, np.pi, 1000)

plt.figure(figsize=(4,3))

plt.plot(x, xi(x, .2))
plt.plot(x, xi(x, 1), "--")
plt.xlim(-np.pi, np.pi)
plt.ylim(-.5,1.5)
plt.title("Profil d'Ã©nergie potentielle")
plt.xlabel("$\\theta$ (rad)")
plt.ylabel("$\\xi$")
plt.grid()

mysavefig("xi.pdf")
```

```python
eta = 0.2

def Epp(x):
    return -eta*np.cos(x)
def Epe(x):
    return 2 * (np.abs(np.cos(x/2)) - .4)**2

x = np.linspace(-3*np.pi, 3*np.pi, 100000)

plt.plot(x, Epp(x))
plt.plot(x, Epe(x))


plt.plot(x, xi(x, eta))


from scipy.misc import derivative
def f(x):
    return xi(x, eta)

print(derivative(f, np.pi))

```

```python
m = 1
g = 1
k = 1

def xi(x, a=1, l0=1):
    return -m*g/k/a*np.cos(x) + 2 * (np.cos(x/2) - l0/2/a)**2

x = np.linspace(-np.pi, np.pi, 1000)

for a in [.1,.2, .3, 1]:
    ep = xi(x, l0=a)
    plt.plot(x, ep-min(ep), label=a)
plt.legend()
```
